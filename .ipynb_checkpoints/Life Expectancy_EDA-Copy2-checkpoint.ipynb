#### 3. What is the average life expectancy in developing and developed countries each year?

To compute this, we have to group the data frame with status and year.

```python
# sns.lineplot(data=status, x=status['Year'], y=status['Life expectancy '], hue='Status')
status = round(new_data.groupby(['Status','Year'])['Life expectancy '].mean(), 2)
```

```python
status = status.reset_index()
```

```python
status.pivot(index='Year', columns='Status')
```

As expected, life expectancy of developed nations is greater than developing nations in each year.

```python
#comparing the life expectancy trend for developed and developing countries over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=status, x=status['Year'], y=status['Life expectancy '], hue='Status')
```

**Observations**
we can clearly see that there was has be improvement in both developed and developing countries over the years(2000-2015)\
however there was a 
+ decline in 2008 and 2012 for developed countries 
+ decline bewteen 2009 and 2010 for developing countries
+ a slow growth between 2002 and 2004 

### 4. How do the various immunizations(Hepatitis B, Polio and Diphtheria) relate to life expectancy?

```python
# using the correlation and heatmap
# calculate correlation
plt.figure(figsize=(12, 6))
corr = new_data.corr()
# plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns,yticklabels=corr.columns, annot=True, cmap="Spectral")
plt.title("Heatmap showing the correlation between variables", fontdict = {'fontsize' : 26});
```

**Obervation**
+ There is a weak positive correlation between Life expectancy and Hepatitis B (r=0.2)
+ There is a moderate positive correlation between Life expectancy and Polio (r=0.46)
+ There is a moderate positive correlation between Life expectancy and Diphtheria (r=0.48)
+ There is a stront positive correlation between Life expectancy and Scooling (r=0.75)

so in general we can say that, increase in immunization will improve life expectancy\
and increase in schooling also contribute to improved life expectancy and vice versa
